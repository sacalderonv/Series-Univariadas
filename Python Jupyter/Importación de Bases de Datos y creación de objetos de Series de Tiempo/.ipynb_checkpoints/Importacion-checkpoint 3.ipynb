Como en R, en Python tambi√©n tenemos algunas caracter√≠sticas que pueden poseer las series de tiempo:

Timestamps(Marcas de tiempo): Instantes espec√≠ficos en el tiempo, o,
Periodos F√≠jos: tales como el mes de enero de a√±o 2020 o el a√±o completo 2020, o,
Intervalo de tiempo: indica el inicio y el final de las marcas de tiempo,
Tiempo transcurrido: cada marca de tiempo es una medida de tiempo relativa a una hora de inicio en particular. Esta secci√≥n fue obtenida del libro "Python for Data Analysis
Data Wrangling with Pandas, NumPy, and IPython " Wes McKinney 2017. Ver el overview en https://pandas.pydata.org/pandas-docs/stable/user_guide/timeseries.html

datetime:  Stores both date and time

timedelta: Represents the difference between two datetime values (as days, seconds, and microseconds) 

tzinfo:    Base type for storing time zone information

## Rangos de datos, frecuencias y saltos

En general, las series son asumidas como irregulares, es decir, no tiene frecuencia fija. Pero hay ocasiones en las cuales se desea tener una frecuencia fija, por ejemplo diaria, mensual o cada 15 minutos, etc. 
Pandas permite trabajar con series de tiempo de frecuencia fija. Es decir, se puede remuestrear(convertir una serie de un frecuencia a otra), inferir la fecuencia, y generar rangos de fechas con
frecuencia fija.

Convertir una serie de alta frecuencia a una de baja frecuencia se llama downsampling, an√°logamente, convertir una serie de baja frecuencia en una de alta se llama upsampling. Sin embargo no todo tipo de remuestreo cae en estas dos categor√≠as. Veamos un ejemplo:

```python
rng = pd.date_range('2000-01-01', periods=100, freq='D')
ts = pd.Series(np.random.randn(len(rng)), index=rng)
ts  ##Esta serie tiene frecuencia diaria.
```

Note por ejemplo que vamos a muestrear o agrupar las observaciones por meses 'M' y luego aplicamos la media,

```python
ts.resample('M').mean()   
```

```python
from statistics import mean
mean(ts['1/1/2000':'31/1/2000'])
```

```python
ts.resample('M', kind='period').mean()  #No hay diferencia, lo √∫nico es que le pone el mes en el √≠ndice
```

## Downsampling

Vamos a convertir una serie de frecuencia alta a una de baja. Es decir, pasaremos de una frecuencia de minutos a 5 minutos

```python
rng = pd.date_range('2000-01-01', periods=12, freq='T')
ts = pd.Series(np.arange(12), index=rng)
ts
```

```python
ts.resample('5min', closed='right').sum()
```
