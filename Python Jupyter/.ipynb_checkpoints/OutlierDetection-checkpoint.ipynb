```python
from sklearn.metrics import mean_squared_error
```

```python
serie=SerieRetColCap
sizetrain=int(len(serie)*0.80)
sizetrain
train,test=serie[0:(sizetrain)],serie[(sizetrain):(len(serie))]
history=[x for x in train]
predicciones=list()
for t in range(len(test)):
    modelo=sm.tsa.statespace.SARIMAX(history, trend='n', order=(0,0,ma_orden))
    modelo_fit=modelo.fit(disp=0)
    output=modelo_fit.forecast()
    yhat=output[0]
    predicciones.append(yhat)
    obs=test[t]
    history.append(obs)
    print('Predicción=%f, Esperado=%f' % (yhat,obs))

error=mean_squared_error(test,predicciones)
print('ECM: %.10f' % error)

```
